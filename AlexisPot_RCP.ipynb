```python
#Parameters
rc=2.5
r0=1.5
```

```python
def alexispot(r0,rc,n):
    r=np.linspace(1, rc, n)
    u=np.random.rand(n)
    eps=1.0/3.0
    lam=1.717
    om=5.0
    sig=1.0
    for i in range(n):
        if(r[i]<rc):    
            phi=(4.0*eps)*(((r[i]/sig)**-12)-((r[i]/sig)**-6)) #the standard LJ
            Gc= lam*eps*(exp(-((om*(r[i]-r0))/(sig))**2))#Gaussian contribution
            phirc=(4.0*eps)*((rc/sig)**-12-(rc/sig)**-6)
            u[i]=phi - Gc -phirc
        else:
            u[i]=0.0
            
    fig=plt.figure()
    ax=fig.add_axes([0,0,1.5,1.5])
    ax.plot(r,u)
    
```

```python
eps=[0.33,0.38,0.41,0.42,0.43]
alexispot(r0,rc,1000)
```

```python
def alexispoteps_r(eps,r0,rc,n):
    r=np.linspace(1, rc, n)
    u=np.random.rand(n)
    lam=1.717
    om=5.0
    sig=1.0
    for i in range(n):
        if(r[i]<rc):    
            phi=(4.0*eps)*(((r[i]/sig)**-12)-((r[i]/sig)**-6)) #the standard LJ
            Gc= lam*eps*(exp(-((om*(r[i]-r0))/(sig))**2))#Gaussian contribution
            phirc=(4.0*eps)*((rc/sig)**-12-(rc/sig)**-6)
            u[i]=phi - Gc -phirc
            
        else:
                u[i]=0.0
    return r

def alexispoteps_u(eps,r0,rc,n):
    r=np.linspace(1, rc, n)
    u=np.random.rand(n)
    lam=1.717
    om=5.0
    sig=1.0
    for i in range(n):
        if(r[i]<rc):    
            phi=(4.0*eps)*(((r[i]/sig)**-12)-((r[i]/sig)**-6)) #the standard LJ
            Gc= lam*eps*(exp(-((om*(r[i]-r0))/(sig))**2))#Gaussian contribution
            phirc=(4.0*eps)*((rc/sig)**-12-(rc/sig)**-6)
            u[i]=phi - Gc -phirc
            
        else:
                u[i]=0.0
    return u
```

```python
r25, u25=alexispoteps_r(0.4,r0,rc,1000),alexispoteps_u(0.4,r0,rc,1000)
r20, u20 = alexispoteps_r(0.41,r0,rc,1000), alexispoteps_u(0.41,r0,rc,1000)
r40, u40 =alexispoteps_r(0.38,r0,rc,1000),alexispoteps_u(0.38,r0,rc,1000)
r5, u5=alexispoteps_r(0.43,r0,rc,1000), alexispoteps_u(0.43,r0,rc,1000)

fig=plt.figure()
ax=fig.add_axes([0,0,1.5,1.5])
ax.plot(r5,u5,color='magenta',label='T=5')
ax.plot(r20,u20,color='violet',label='T=20')
ax.plot(r25,u25,color='green',label='T=25')
ax.plot(r40,u40,color='blue',label='T=40')
plt.legend()
```
